```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Link = 'https://github.com/seismomat/HerramientasComputacionales/blob/main/Aplicaciones/ebola_outbreak_dataset-master/ebola_outbreak_dataset-master/ebola_2014_2016_clean.csv'
code = '?raw=true'
url = Link+code
df = pd.read_csv(url,sep=',',index_col=1)
```

```python
NI=df[df['Country']=='Nigeria'] #Creamos variable de Nigeria
```

```python
NI #Imprimimos datos sobre nigeria
```

```python
NI = NI.drop('Country',axis=1)
```

```python
Dates = NI.index[0:-1:10] #Creamos variables diferentes
NCC=NI['No. of confirmed cases'][0:-1:10]
NCD=NI['No. of confirmed deaths'][0:-1:10]
```

```python
plt.plot(Dates,NCC,'o--m',ms= 4,label='Infectados')
plt.bar(Dates,NCC, color = 'magenta', alpha= 0.2) #Añade barras
plt.plot(Dates,NCD,'d--b',ms= 4,label='Decesos')
plt.bar(Dates,NCD, color = 'blue', alpha= 0.2) #Añade barras
plt.grid() #Crea una cuadricula en nuestra gráfica
plt.title('Avance del Ébola en Nigeria') #Titulo principal
plt.ylabel('Casos Confirmados') #Da un titulo a cada eje
plt.xlabel('Perdiodo de tiempo') #Da un titulo a cada eje
plt.xticks(rotation=90,fontsize=8) #Acomodamos las variables del eje x
plt.legend() #Cuadrito de información
plt.savefig("Barras_NI.png", dpi = 300,format = 'png') #Datos para cuardar nuestra figura
plt.show()
```

```python
PIvsD = 100*NI['No. of confirmed deaths']/NI['No. of confirmed cases'] #Nos da un contraste entre variables
PIvsD
```

```python
NI['PIvsD']=PIvsD
```

```python
recuperados = ((100-PIvsD)/100)*NI['No. of confirmed cases'] #Promedio de variables para crear una nueva variable
NI['recuperados'] = recuperados
```

```python
NI['PR'] = (100-PIvsD) #Nueva variable sobre recuperados
```

```python
NI.head() #Nos da el top 5 de nuestros datos mas las variables que hemos definido anteriormente
```

```python
NI.tail()#Nos da el bottom 5 de nuestros datos mas las variables que hemos definido anteriormente
```

```python
PIvsD_mean = NI['PIvsD'].mean() #Promedio de nuestra variable PIvsD
PR_mean = NI['PR'].mean() #Promedio de nuestra variable PR
porcentajes = [PIvsD_mean,PR_mean] #definimos variables
etiquetas = ['PIvsD','PR']
```

```python
plt.pie(porcentajes,labels=etiquetas,autopct='%0.1f %%', colors=['gray','pink']) #grafica pastel con sus datos
plt.axis("equal") #Definimos al eje X y X del mismo tamaño
plt.title('Porcentaje entre recuperados vs decesos') #titulo principal
plt.savefig("Pastel_PIvsD_PR", dpi = 300,format = 'png') #datos de guardado
plt.show()
```

```python
Ñ = NI['Girls'][0:-1:16] #Grafica de datos con la variable de Nigeria
MG = NI['Adult Female'][0:-1:16]
MV = NI['Old Female'][0:-1:16]
Dates = NI.index[0:-1:16]

fig = plt.figure(figsize=(12,6))
plt.bar(Dates,Ñ,color='red',alpha=0.6,label='Niñas') #Generamos barras para cada variable
plt.bar(Dates,MG,color='gray',alpha=0.4,label='Mujeres Adultas')
plt.bar(Dates,MV,color='magenta',alpha=0.4,label='Mujeres Mayores')
plt.grid() #Cuadricula en gráfica
plt.title('Distribución del Ébola en Mujeres') #titulo principal
plt.ylabel('Casos Confirmados') #titulo del eje y
plt.xticks(rotation=90,fontsize=8) #Nos ayuda a acomodar nuestros datos
plt.legend(loc='upper left') #crea cuadrito de información
plt.savefig('FemaleNI.png',dpi = 300,format = 'png') #datos de guardado
plt.show()
```
